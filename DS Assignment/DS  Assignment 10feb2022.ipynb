# Feature Engineering
Add Column




# High_Low_HP

```python
hp_mean =data["HP"].mean()

def hp_set(val):
    if val < hp_mean:
        return "HP Low"
    elif val == hp_mean:
        return "HP Netural"
    else:
        return "HP High"
```

```python
# "HP_high_low"  -> "HP High" , "HP Low"
data ["HP_high_low"]= data["HP"].apply(hp_set)
data
```

# High_Low_Speed

```python
data["HP"].mean()
```

```python
speed_mean =data["HP"].mean()

def speed_set(val):
    if val < speed_mean:
        return "speed Low"
    elif val == speed_mean:
        return "speed Netural"
    else:
        return "speed High"
```

```python
# "Speed_high_low"  -> "Speed High" , "Speed Low"
data ["Speed_high_low"]= data["Speed"].apply(speed_set)
data
```
